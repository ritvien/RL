# Reinforcement Learning

* **Environment** (Môi trường): là không gian mà máy tương tác
* **Agent** (Máy): là chủ thể tương tác với môi trường qua hành động
* **Policy** (Chiến thuật): là chiến thuật mà máy sử dụng để đưa ra hành động
* **State** (Trạng thái): mô tả trạng thái hiện tại của máy
* **Reward** (Phần thưởng): phần thưởng từ môi trường tương ứng với hành động được thực hiện
* **Action** (Hành động): là những gì máy có thể thực hiện

<font size =5> &#x1F9EE; VD: 

👉 bắt đầu từ state **S(t)**, tại đây, `agent` sẽ dựa trên `policy` đã được thiết kế sẵn để đưa ra 1 `action` trong `environment`

👉 Điều này làm `environment` chuyển đổi sang trạng thái tiếp theo: **S(t+1)** và đưa ra phần thưởng **R(t)** tương ứng cho `action` vừa thực hiện

👉`Agent` sẽ lặp đi lặp lại quá trình này đến khi tìm được chuỗi các hành động tối đa hóa phần thưởng được nhận

## Q-table

**Cột states**: Đại diện cho bất kỳ 1 state nào có thể xảy ra

**Các cột action**: Thể hiện độ ưu tiên(nên được thực hiện) của một `action` nào đấy khi đang ở vị trí `state` tương ứng

mỗi giá trị trong `cell` của `Q-table` là thứ ta cần cập nhật sau mỗi lần thực hiện action, và training để thay đổi độ ưu tiên thực hiện các **action** khác nhau

<font size =5> &#x1F9EE; **Công thức cập nhật Q**:

<span style="white-space: nowrap;"> <font size=5> $Q^{new}(s_t,a_t) \leftarrow $<span style="color: red;">$(1-\alpha).Q(s_t,a_t)$ </span> $+ $<span style="color: green;">$\alpha . (r_t+\gamma.  \max_{\alpha}Q(s_{t+1},a))$ </span> 

<div style="margin-left: 200px;"><span style="color: red;">Thành phần giữ lại</span> 


<div style="margin-left: 450px;"><span style="color: green;">Thành phần cập nhật</span> 

⚠️ Trong đó:

* $s_t$ là state tại thời điểm t
* $\alpha$ là `learning rate`
* $a$ là action
* $r_t$ là `reward` sau khi thực hiện `action` $a_t$
* $\gamma$ là **`discount factor`**
